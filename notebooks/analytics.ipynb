Extração do Dataset Macro do python-bcb

```python
# Imports
from bcb import sgs
import pandas as pd
from datetime import date, timedelta

# Período: últimos 10 anos
data_fim = date.today()
data_inicio = date(data_fim.year - 10, data_fim.month, data_fim.day)

print(f"Período: {data_inicio} até {data_fim}")
```

```python
# Download dos dados
# Códigos SGS do Banco Central:
# 432  = IPCA (% ao mês)
# 11   = Selic (% ao ano)
# 1    = Dólar (taxa de câmbio - venda)

dados = sgs.get(
    codes={
        'selic': 11,
        'ipca': 433,
        'dolar': 1
    },
    start=data_inicio,
    end=data_fim
)

dados.index.name = 'data'
print(dados.shape)
dados.head()
```

```python
# Salvar CSV
dados.to_csv('indicadores_economicos_10anos.csv', sep=';', decimal=',', encoding='utf-8-sig')
print("✅ Arquivo salvo: indicadores_economicos_10anos.csv")
dados.describe()
```
